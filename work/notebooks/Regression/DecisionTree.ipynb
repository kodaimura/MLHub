```julia
using DecisionTree
using Plots
```

```julia
X = rand(20)
Y = 4 .* X .+ 1 .+ 0.2 .* randn(20)
```

```julia
X_matrix = reshape(X, :, 1) 
```

```julia
model = DecisionTreeRegressor(max_depth=5, min_samples_leaf=2)
```

```julia
fit!(model, X_matrix, Y)
```

```julia
X_test = reshape(collect(range(0, 1, length=100)), :, 1)
Y_pred = predict(model, X_test)
```

```julia
scatter(X, Y, label="Training Data", legend=:topleft)
plot!(X_test[:], Y_pred, label="Decision Tree Regression", linewidth=2)
```

```julia
print_tree(model)
```

```julia
X3 = rand(20, 2)
Y3 = sin.(2 * π .* X3[:, 1]) .+ cos.(2 * π .* X3[:, 2]) .+ 0.1 .* randn(20)
```

```julia
model3 = DecisionTreeRegressor(max_depth=5, min_samples_leaf=2)
```

```julia
fit!(model3, X3, Y3)
```

```julia
X_test3 = rand(100, 2)
Y_pred3 = predict(model3, X_test3)
```

```julia
scatter3d(X3[:, 1], X3[:, 2], Y3, label="Data points", xlabel="Feature 1", ylabel="Feature 2", zlabel="Target", legend=:topright)
scatter3d!(X3[:, 1], X3[:, 2], Y_pred3, label="Predictions", color=:orange)
```
